1. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ —Ç–∏–ø—ã –≥—Ä–∞—Ñ–∏–∫–æ–≤ (–ø—Ä–æ—Å—Ç—ã–µ, —Ç–æ—á–µ—á–Ω—ã–µ, –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∏ —Ç.–¥.) —Ñ—É–Ω–∫—Ü–∏–∏ $ùë¶ = sin(ùë•)$, $ùë• = 0, 2ùúã$. –û—Ç–æ–±—Ä–∞–∑–∏—Ç–µ –≤—Å–µ –≥—Ä–∞—Ñ–∏–∫–∏ –≤ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–º –æ–∫–Ω–µ.

```julia
using Plots

# –ó–∞–¥–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è x –æ—Ç 0 –¥–æ 2œÄ
x = range(0, stop=2œÄ, length=100)
n = 1

# –í—ã—á–∏—Å–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è y = sin(x)
y = sin.(x)

# –°–æ–∑–¥–∞–µ–º –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–µ –æ–∫–Ω–æ —Å 2 —Å—Ç—Ä–æ–∫–∞–º–∏ –∏ 2 —Å—Ç–æ–ª–±—Ü–∞–º–∏
plot_layout = @layout [a b; c d]

# –°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏
p1 = plot(x, y, label="sin(x)", xlabel="x", ylabel="y", legend=:topleft)

# –°—Ç—Ä–æ–∏–º —Ç–æ—á–µ—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫
p2 = scatter(x, y, label="sin(x)", xlabel="x", ylabel="y", legend=:topleft)

# –°—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É
p3 = histogram(n, y, label="sin(x)", xlabel="y", ylabel="Frequency", legend=:topleft)

# –°—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É
p4 = bar(x, y, label="sin(x)", xlabel="x", ylabel="y", legend=:topleft)

# –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≤—Å–µ –≥—Ä–∞—Ñ–∏–∫–∏ –≤ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–º –æ–∫–Ω–µ
plot(p1, p2, p3, p4, layout=plot_layout)
```

2. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–∏ $ùë¶ = sin(ùë•)$, $ùë• = 0, 2ùúã$ —Å–æ –≤—Å–µ–º–∏ –≤–æ–∑–º–æ–∂–Ω—ã–º–∏ (—Å–∫–æ–ª—å–∫–æ —Å–º–æ–∂–µ—Ç–µ –≤—Å–ø–æ–º–Ω–∏—Ç—å) —Ç–∏–ø–∞–º–∏ –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏—è –ª–∏–Ω–∏–π –≥—Ä–∞—Ñ–∏–∫–∞. –û—Ç–æ–±—Ä–∞–∑–∏—Ç–µ –≤—Å–µ –≥—Ä–∞—Ñ–∏–∫–∏ –≤ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–º –æ–∫–Ω–µ.

```julia
using Plots

x = 0:0.01:2œÄ
y = sin.(x)

linestyles = [:solid, :dot, :dash, :dashdot, :dashdotdot]

plot(x, y, label="sin(x)", linewidth=2, legend=:topleft) # –ù–∞—á–∞–ª—å–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ —Å –æ–±—ã—á–Ω–æ–π –ª–∏–Ω–∏–µ–π

for (i, linestyle) in enumerate(linestyles)
    plot!(x .+ i*0.05, y, label="Line $i", linestyle=linestyle, linewidth=2) # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Å —Ä–∞–∑–Ω—ã–º–∏ —Ç–∏–ø–∞–º–∏ –ª–∏–Ω–∏–π
end

display(plot!()) # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≤—Å–µ—Ö –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ –æ–¥–Ω–æ–º –æ–∫–Ω–µ
```

3. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ $ùë¶(ùë•) = \pi x^{2}\ln(x)$, –Ω–∞–∑–æ–≤–∏—Ç–µ –æ—Å–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –ü—É—Å—Ç—å —Ü–≤–µ—Ç —Ä–∞–º–∫–∏ –±—É–¥–µ—Ç –∑–µ–ª—ë–Ω—ã–º, –∞ —Ü–≤–µ—Ç —Å–∞–º–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ ‚Äî –∫—Ä–∞—Å–Ω—ã–º. –ó–∞–¥–∞–π—Ç–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –Ω–∞–¥–ø–∏—Å—è–º–∏ –∏ –æ—Å—è–º–∏ —Ç–∞–∫, —á—Ç–æ–±—ã –Ω–∞–¥–ø–∏—Å–∏ –ø–æ–ª–Ω–æ—Å—Ç—å—é —É–º–µ—â–∞–ª–∏—Å—å –≤ –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–º –æ–∫–Ω–µ. –ó–∞–¥–∞–π—Ç–µ —à—Ä–∏—Ñ—Ç –Ω–∞–¥–ø–∏—Å–µ–π. –ó–∞–¥–∞–π—Ç–µ —á–∞—Å—Ç–æ—Ç—É –æ—Ç–º–µ—Ç–æ–∫ –Ω–∞ –æ—Å—è—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç.

```julia
using Plots

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏
y1(x) = pi * x^2 * log(x)

# –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –∑–Ω–∞—á–µ–Ω–∏–π x
x_values = 0.1:0.1:2.5

# –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –∑–Ω–∞—á–µ–Ω–∏–π y
y_values = y1.(x_values)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plot(x_values, y_values, color=:red, xlabel="x", ylabel="y", label="\$y = \\pi x^{2} \\ln(x)\$", 
     title="–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏", linewidth=2, legend=:topright,
     framestyle=:box, grid=:on, size=(600, 400),
     guidefont=font(12, "Arial"), tickfont=font(10, "Arial"))

display(plot!())  # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
```

4. –ó–∞–¥–∞–π—Ç–µ –≤–µ–∫—Ç–æ—Ä $ùë• = (‚àí2, ‚àí1, 0, 1, 2)$. –í –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–º –æ–∫–Ω–µ (–≤ 4-—Ö –ø–æ–¥–æ–∫–Ω–∞—Ö) –∏–∑–æ–±—Ä–∞–∑–∏—Ç–µ –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –ø–æ —Ç–æ—á–∫–∞–º ùë• –∑–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ $ùë¶(ùë•)$ = $x^{3} ‚àí 3ùë•$ –≤ –≤–∏–¥–µ:\
‚Äì —Ç–æ—á–µ–∫,\
‚Äì –ª–∏–Ω–∏–π,\
‚Äì –ª–∏–Ω–∏–π –∏ —Ç–æ—á–µ–∫,\
‚Äì –∫—Ä–∏–≤–æ–π.\
–°–æ—Ö—Ä–∞–Ω–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ —Ñ–∞–π–ª–µ figure_familiya.png, –≥–¥–µ –≤–º–µ—Å—Ç–æ familiya —É–∫–∞–∂–∏—Ç–µ –≤–∞—à—É —Ñ–∞–º–∏–ª–∏—é

```julia
using Plots

# –ó–∞–¥–∞–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–∞ x
x = [-2, -1, 0, 1, 2]

# –§—É–Ω–∫—Ü–∏—è y(x) = x^3 - 3x
y2(x) = x^3 - 3x

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≤ 4-—Ö –ø–æ–¥–æ–∫–Ω–∞—Ö
plot1 = scatter(x, y2.(x), markersize=4, markercolor=:blue, xlabel="x", ylabel="y", label="–¢–æ—á–∫–∏")
plot2 = plot(x, y2.(x), line=:line, linewidth=2, linecolor=:green, xlabel="x", ylabel="y", label="–õ–∏–Ω–∏–∏")
plot3 = plot(x, y2.(x), line=:line, linewidth=2, linecolor=:orange, markersize=4, markercolor=:red, m=:circle, xlabel="x", ylabel="y", label="–õ–∏–Ω–∏–∏ –∏ —Ç–æ—á–∫–∏")
plot4 = curves(x, y2.(x), line=:auto, linewidth=2, linecolor=:purple, xlabel="x", ylabel="y", label="–ö—Ä–∏–≤–∞—è")

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ –æ–¥–Ω–æ –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–µ –æ–∫–Ω–æ
plot(plot1, plot2, plot3, plot4, layout=(2,2), legend=:bottomright, size=(800, 600))

display(plot!())
```

5. –ó–∞–¥–∞–π—Ç–µ –≤–µ–∫—Ç–æ—Ä $x = (3, 3.1, 3.2, ‚Ä¶ , 6)$. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–π $y_{1}\left(x\right)$ = $\pi x$ –∏ $y_{2}\left(x\right)$ = $\exp\left(x\right)\cos\left(x\right)$ –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–º –¥–∏–∞–ø–∞–∑–æ–Ω–µ –∑–Ω–∞—á–µ–Ω–∏–π –∞—Ä–≥—É–º–µ–Ω—Ç–∞ ùë• —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:\
‚Äì –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –æ–±–∞ –≥—Ä–∞—Ñ–∏–∫–∞ —Ä–∞–∑–Ω–æ–≥–æ —Ü–≤–µ—Ç–∞ –Ω–∞ –æ–¥–Ω–æ–º —Ä–∏—Å—É–Ω–∫–µ, –¥–æ–±–∞–≤—å—Ç–µ –ª–µ–≥–µ–Ω–¥—É –∏ —Å–µ—Ç–∫—É –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞; —É–∫–∞–∂–∏—Ç–µ –Ω–µ–¥–æ—Å—Ç–∞—Ç–∫–∏ —É –¥–∞–Ω–Ω–æ–≥–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è;\
‚Äì –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –∞–Ω–∞–ª–æ–≥–∏—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ —Å –¥–≤—É–º—è –æ—Å—è–º–∏ –æ—Ä–¥–∏–Ω–∞—Ç.


```julia
using Plots

# –ó–∞–¥–∞–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–∞ x
x = 3:0.01:6

# –§—É–Ω–∫—Ü–∏–∏ y1(x) –∏ y2(x)
y11(x) = œÄ * x
y22(x) = exp(x) * cos(x)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –Ω–∞ –æ–¥–Ω–æ–º —Ä–∏—Å—É–Ω–∫–µ —Å –ª–µ–≥–µ–Ω–¥–æ–π –∏ —Å–µ—Ç–∫–æ–π
plot(x, y1.(x), label="y‚ÇÅ(x) = œÄx", color=:blue, xlabel="x", ylabel="y‚ÇÅ(x)",
     title="–ì—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–π", linewidth=2, legend=:left, grid=:on)
plot!(x, y2.(x), label="y‚ÇÇ(x) = eÀ£cos(x)", color=:red, linewidth=2)

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
display(plot!())
```

```julia
using Plots

# –ó–∞–¥–∞–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–∞ x
x = 3:0.1:6

# –§—É–Ω–∫—Ü–∏–∏ y1(x) –∏ y2(x)
y11(x) = œÄ * x
y22(x) = exp(x) * cos(x)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ —Å –¥–≤—É–º—è –æ—Å—è–º–∏ –æ—Ä–¥–∏–Ω–∞—Ç
plot(x, y1.(x), label="y‚ÇÅ(x) = œÄx", color=:blue, xlabel="x", ylabel="y‚ÇÅ(x)",
     title="–ì—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–π —Å –¥–≤—É–º—è –æ—Å—è–º–∏ –æ—Ä–¥–∏–Ω–∞—Ç", linewidth=2, legend=:left, grid=:on)
plot!(twinx(), x, y2.(x), secondary=true, label="y‚ÇÇ(x) = eÀ£cos(x)", color=:red, linewidth=2,
      ylabel="y‚ÇÇ(x)", linestyle=:dot)

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
display(plot!())
```

6. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (–ø—Ä–∏–¥—É–º–∞–π—Ç–µ —Å–∞–º–∏), —É—á–∏—Ç—ã–≤–∞—è –æ—à–∏–±–∫—É –∏–∑–º–µ—Ä–µ–Ω–∏—è.

9. –°–æ–∑–¥–∞–π—Ç–µ –∞–Ω–∏–º–∞—Ü–∏—é —Å –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ–º —Å–∏–Ω—É—Å–æ–∏–¥—ã. –¢–æ –µ—Å—Ç—å –≤—ã —Å—Ç—Ä–æ–∏—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å 
–≥—Ä–∞—Ñ–∏–∫–æ–≤ —Å–∏–Ω—É—Å–æ–∏–¥—ã, –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ —É–≤–µ–ª–∏—á–∏–≤–∞—è –∑–Ω–∞—á–µ–Ω–∏–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞. –ü–æ—Å–ª–µ —Å–æ–µ–¥–∏–Ω–∏—Ç 
–∏—Ö –≤ –∞–Ω–∏–º–∞—Ü–∏—é.

```julia
using Plots

# –°–æ–∑–¥–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏, –∫–æ—Ç–æ—Ä—É—é –±—É–¥–µ–º –∞–Ω–∏–º–∏—Ä–æ–≤–∞—Ç—å
f(x, œâ, t) = sin(œâ * x + t)

# –î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π x
x_values = 0:0.1:10

# –î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π –≤—Ä–µ–º–µ–Ω–∏ –¥–ª—è –∞–Ω–∏–º–∞—Ü–∏–∏
time_values = 0:0.1:2œÄ

# –°–æ–∑–¥–∞–Ω–∏–µ –∞–Ω–∏–º–∞—Ü–∏–∏
anim = @animate for t in time_values
    y_values = f.(x_values, 1, t)  # –ß–∞—Å—Ç–æ—Ç–∞ œâ = 1
    plot(x_values, y_values, label="–°–∏–Ω—É—Å–æ–∏–¥–∞", xlabel="x", ylabel="y", title="–ê–Ω–∏–º–∞—Ü–∏—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Å–∏–Ω—É—Å–æ–∏–¥—ã",
         ylims=(-1.5, 1.5), linewidth=2, color=:blue)
end

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –∞–Ω–∏–º–∞—Ü–∏–∏ –≤ —Ñ–∞–π–ª
gif(anim, "sinusoid_animation.gif", fps = 10)
```

10. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –∞–Ω–∏–º–∏—Ä–æ–≤–∞–Ω–Ω—É—é –≥–∏–ø–æ—Ü–∏–∫–ª–æ–∏–¥—É –¥–ª—è $2$ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–æ–¥—É–ª—è $ùëò$ –∏ $2$ —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–æ–¥—É–ª—è $ùëò$.

```julia
using Plots

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –≥–∏–ø–æ—Ü–∏–∫–ª–æ–∏–¥—ã
function hypocycloid(a, b, k)
    Œ∏ = range(0, stop=2œÄ, length=100)
    x = (a - b) * cos.(Œ∏) .+ b * k * cos.((a - b) / b * Œ∏)
    y = (a - b) * sin.(Œ∏) .- b * k * sin.((a - b) / b * Œ∏)
    return x, y
end

# –ó–Ω–∞—á–µ–Ω–∏—è –º–æ–¥—É–ª—è k (–¥–ª—è —Ü–µ–ª—ã—Ö –∏ —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª)
k_values = [2, 3//2, 4, 5//2]

# –°–æ–∑–¥–∞–Ω–∏–µ –∞–Ω–∏–º–∞—Ü–∏–∏ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π k
anim = @animate for k in k_values
    x, y = hypocycloid(1, 0.3, k)
    
    plot(x, y, aspect_ratio=:equal, label="k = $k", xlabel="x", ylabel="y",
         title="–ê–Ω–∏–º–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –≥–∏–ø–æ—Ü–∏–∫–ª–æ–∏–¥–∞", linewidth=2)
end

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –∞–Ω–∏–º–∞—Ü–∏–∏ –≤ —Ñ–∞–π–ª
gif(anim, "hypocycloid_animation.gif", fps=5)
```

11. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –∞–Ω–∏–º–∏—Ä–æ–≤–∞–Ω–Ω—É—é —ç–ø–∏—Ü–∏–∫–ª–æ–∏–¥—É –¥–ª—è $2$ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–æ–¥—É–ª—è $ùëò$ –∏ $2$ —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–æ–¥—É–ª—è $ùëò$.

```julia
using Plots

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —ç–ø–∏—Ü–∏–∫–ª–æ–∏–¥—ã
function epicycloid(a, b, k)
    Œ∏ = range(0, stop=2œÄ, length=100)
    x = (a + b) * cos.(Œ∏) .- b * k * cos.((a + b) / b * Œ∏)
    y = (a + b) * sin.(Œ∏) .- b * k * sin.((a + b) / b * Œ∏)
    return x, y
end

# –ó–Ω–∞—á–µ–Ω–∏—è –º–æ–¥—É–ª—è k (–¥–ª—è —Ü–µ–ª—ã—Ö –∏ —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª)
k_values = [2, 3//2, 4, 5//2]

# –°–æ–∑–¥–∞–Ω–∏–µ –∞–Ω–∏–º–∞—Ü–∏–∏ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π k
anim = @animate for k in k_values
    x, y = epicycloid(1, 0.3, k)
    
    plot(x, y, aspect_ratio=:equal, label="k = $k", xlabel="x", ylabel="y",
         title="–ê–Ω–∏–º–∏—Ä–æ–≤–∞–Ω–Ω–∞—è —ç–ø–∏—Ü–∏–∫–ª–æ–∏–¥–∞", linewidth=2)
end

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –∞–Ω–∏–º–∞—Ü–∏–∏ –≤ —Ñ–∞–π–ª
gif(anim, "epicycloid_animation.gif", fps=5)
```
